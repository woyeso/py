### Why Numpy array is useful

```python
SIZE = 100000
a = range(SIZE)
b = range(SIZE)
```

```python
# python list
start = time.time()
result = [(x+y) for x,y in zip(a,b)]
print("python list took: ",(time.time()-start)*1000)
```

```python
# numpy array
a1=np.arange(SIZE)
a2=np.arange(SIZE)
```

```python
start= time.time()
result = a1 + a2
print("numpy took: ", (time.time()-start)*1000)
```
